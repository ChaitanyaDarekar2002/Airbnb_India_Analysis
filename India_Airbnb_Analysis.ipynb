```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Airbnb_India_Top_500.csv")
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['stars']=data['stars'].fillna(data['stars'].mean())
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data['stars'] = data['stars'].round(2)
```

```python
data.head()
```

```python
def price_category(x):
    if x < 5000:
        return 'Budget'
    elif x < 15000:
        return 'Mid-Range'
    else:
        return 'Luxury'


data['Price_Category'] = data['pricing/rate/amount'].apply(price_category)
```

```python
print(data.groupby('roomType')['pricing/rate/amount'].mean().round(2))
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['pricing/rate/amount'])
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
data['address'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Airbnb Locations')
plt.show()
```

```python
data.groupby('roomType')['pricing/rate/amount'].mean().head(10).sort_values().plot(kind='barh',color='red')
plt.title('Average Price by Room Type')
plt.show()
```

```python
plt.scatter(data['stars'], data['pricing/rate/amount'],color='red')
plt.title('Rating vs Price')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

```python
price_counts = data['Price_Category'].value_counts()
plt.figure(figsize=(7,7))

plt.pie(
    price_counts,
    labels=price_counts.index,
    autopct='%1.1f%%'
)
plt.title('Price Category Distribution')
plt.show()
```
